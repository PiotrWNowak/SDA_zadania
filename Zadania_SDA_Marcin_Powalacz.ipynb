## Zadanie 1: 
Stwórz dataframe z dzięsięcioma imionami uczniów i uczennic oraz liczbą punktów, jakie uzyskali z egzaminu. Następnie sprawdź, jaka była średnia i mediana wyników.

```python
import pandas as pd
import numpy as np
```

###Lista uczniów z wynikami

```python
student_list = pd.DataFrame([['Karolina', 42], 
                             ['Marcin', 48], 
                             ['Krzystof', 25], 
                             ['Aleksandra', 11], 
                             ['Sebastian', 15], 
                             ['Katarzyna', 16], 
                             ['Roman', 41], 
                             ['Marek', 37], 
                             ['Adam', 26], 
                             ['Zenon', 25]],
                            columns=('Name', 'Points'))
student_list
```

###Średnia

```python
from numpy.ma.core import mean
mean = pd.DataFrame.mean(student_list, numeric_only=True)
mean
```

###Mediana

```python
pd.DataFrame.median(student_list, numeric_only=True)
```

##Zadanie 3:
Dodaj do ramki trzecią kolumnę, w której znajdzie się procentowy wynik
uczniów i uczennic z tego egzaminu (to znaczy, że każda wartość liczbowa musi
zostać podzielona przez maksimum punktów, jakie można było na tym
egzaminie uzyskać).
Następnie, chcemy zanonimizować kolumnę z imionami: chcemy, by została
tylko pierwsza i ostatnia litera imienia. Np. dla imienia “Marcelina” chcemy
zachować “M...a”

###Procentowy wynik

```python
percent = student_list['Points'].apply(lambda x: x / 50 * 100)
student_list['%'] = percent
student_list
```

###ukrycie imion

```python
hiden = []
for x in student_list['Name']:
  hiden.append(f'{x[0]}.....{x[-1]}')

hiden
```

```python
student_list['Name'] = hiden
student_list
```
